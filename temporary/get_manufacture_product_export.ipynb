```python
import numpy as np
import pandas as pd  
import time
```

```python
dfUser = pd.read_csv('../data_csv/User.csv')
dfOrderDetail = pd.read_csv('../data_csv/OrderDetail.csv')
dfOrderDetail = dfOrderDetail.rename(columns= {"code" : 'code_order' , 'name' : 'name_order'})[['code_order' , 'name_order' , 'order_id']]
dfManufactureDeptDetail = pd.read_csv('../data_csv/ManufactureDeptDetail.csv')
dfManufactureDeptDetail = dfManufactureDeptDetail.rename(columns= {'code' : 'code_dept' , 'name' : 'name_dept'})
dfManufactureTaskDetail = pd.read_csv('../data_csv/ManufactureTaskDetail.csv')
```

```python
dfManufactureProd = pd.read_csv('../data_csv/ManufactureProd.csv')
dfManufactureProd = dfManufactureProd.rename(columns= {'id' : 'manufacture_product_id' , 'created_at' : 'create_manu_prod' , 'quantity' : 'quantity_manu_prod' , 'name' : 'name_manu_prod'})
dfManufactureProd = dfManufactureProd.drop(columns= {'updated_at','deleted_at' , 'qrcode'})
dfManufactureProd[dfManufactureProd['task_id'] == 1663]
```

```python
# dfManufactureProdExport = pd.read_csv('../data_csv/ManufactureProdExport.csv')
# dfManufactureProdExportDetail = pd.read_csv('../data_csv/ManufactureProdExportDetail.csv')
```

```python
dfManufactureProdExport = pd.read_csv('../data_csv/ManufactureProdExport.csv')
dfManufactureProdExport = dfManufactureProdExport.drop(columns= {'code' , 'deleted_at'})
dfManufactureProdExport[dfManufactureProdExport['task_id'] == 1663]
```

```python
dfManuProdExportMerge = dfManufactureProdExport.merge(dfManufactureProd , on = ['manufacture_product_id' , 'task_id'])
dfManuProdExportMerge = dfManuProdExportMerge.merge(dfOrderDetail , on = 'order_id' , how = 'left')
dfManuProdExportMerge = dfManuProdExportMerge.merge(dfUser , on = 'creator_id' , how = 'left')
dfManuProdExportMerge = dfManuProdExportMerge.merge(dfManufactureDeptDetail[['manufacture_dept_id' , 'name_dept']] , on = 'manufacture_dept_id' , how = 'left')
dfManuProdExportMerge = dfManuProdExportMerge.merge(dfManufactureTaskDetail[['task_id' , 'order_id' , 'start_time' , 'end_time' , 'name_status']])

dfManuProdExportMerge[dfManuProdExportMerge['task_id'] == 1663].loc[4250]
```

```python
dfManuProdExportMergeFix = dfManuProdExportMerge.drop(columns= {'creator_id' , 'created_at' , 'manufacture_dept_id', 'manufacture_product_id' , 'order_id' })

new_col1 = ['task_id', 'code_order' ,'name_order', 'name_dept','fullname', 'updated_at' , 'create_manu_prod' , 'start_time' , 'end_time' ,'quantity_manu_prod' , 
      'quantity', 'position', 'name_status' , 'is_import' , 'receiver']
   
dfManuProdExportMergeFix= dfManuProdExportMergeFix.reindex(columns=new_col1)
   
dfManuProdExportMergeFix.rename(columns = {'code_order': 'Mã đơn hàng' , 'name_order': 'Tên đơn hàng', 'name_dept':'Phòng ban xuất',  'fullname' : 'Người xuất' , 'updated_at': 'Ngày xuất', 'create_manu_prod' : 'Ngày tạo chi tiết công việc','start_time' : 'Ngày BĐ dự kiến' , 'end_time' : 'Ngày kết thúc dự kiến',
                                             'quantity_manu_prod' : 'Số lượng chi tiết CV' , 'quantity' : 'Số lượng xuất' , 'position' : 'Vị trí' , 
                                             'name_status': 'Trạng thái xuất', 'total_prod_quan': 'Tổng số BTP', 'code_status': 'Mã trạng thái', 'name_status' : 'Trạng thái CV' , 'is_import' : 'Đã nhập ?' , 'receiver' : 'Người nhận'}, inplace = True)
dfManuProdExportMergeFix.tail(10)
```

```python
dfManuProdExportDetail = pd.read_csv('../data_csv/ManufactureProdExportDetail.csv')
dfManuProdExportDetail = dfManuProdExportDetail.drop(columns= {'created_at' , 'deleted_at'})
dfManuProdExportDetail[dfManuProdExportDetail['task_id'] == 3]
```

```python
dfManuProdExportDetailMerge = dfManuProdExportDetail.merge(dfManufactureProd , on = ['manufacture_product_id' , 'task_id'])
dfManuProdExportDetailMerge = dfManuProdExportDetailMerge.merge(dfOrderDetail , on = 'order_id' , how = 'left')
dfManuProdExportDetailMerge = dfManuProdExportDetailMerge.merge(dfUser , on = 'creator_id' , how = 'left')
dfManuProdExportDetailMerge = dfManuProdExportDetailMerge.merge(dfManufactureDeptDetail[['manufacture_dept_id' , 'name_dept']] , on = 'manufacture_dept_id' , how = 'left')
dfManuProdExportDetailMerge = dfManuProdExportDetailMerge.merge(dfManufactureTaskDetail[['task_id' , 'order_id' , 'start_time' , 'end_time' , 'name_status']])

dfManuProdExportDetailMerge[dfManuProdExportDetailMerge['task_id'] == 3].head(3)
```

```python
dfManuProdExportDetailMergeFix = dfManuProdExportDetailMerge.drop(columns= {'creator_id' , 'manufacture_dept_id', 'manufacture_product_id' , 'order_id' })

new_col1 = ['task_id', 'code_order' ,'name_order', 'name_dept','fullname', 'updated_at' , 'create_manu_prod' , 'start_time' , 'end_time' ,'quantity_manu_prod' , 
      'quantity', 'position', 'name_status' , 'receiver' ]
   
dfManuProdExportDetailMergeFix= dfManuProdExportDetailMergeFix.reindex(columns=new_col1)
   
dfManuProdExportDetailMergeFix.rename(columns = {'code_order': 'Mã đơn hàng' , 'name_order': 'Tên đơn hàng', 'name_dept':'Phòng ban xuất',  'fullname' : 'Người xuất' , 'updated_at': 'Ngày xuất', 'create_manu_prod' : 'Ngày tạo chi tiết công việc','start_time' : 'Ngày BĐ dự kiến' , 'end_time' : 'Ngày kết thúc dự kiến',
                                             'quantity_manu_prod' : 'Số lượng chi tiết CV' , 'quantity' : 'Số lượng xuất chi tiết' , 'position' : 'Vị trí' , 
                                             'name_status': 'Trạng thái xuất', 'total_prod_quan': 'Tổng số BTP', 'code_status': 'Mã trạng thái', 'name_status' : 'Trạng thái CV' , 'receiver' : 'Người nhận'}, inplace = True)
dfManuProdExportDetailMergeFix.tail(10)
```
