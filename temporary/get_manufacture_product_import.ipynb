```python
import numpy as np
import pandas as pd  
import time
```

```python
dfUser = pd.read_csv('../data_csv/User.csv')
dfOrderDetail = pd.read_csv('../data_csv/OrderDetail.csv')
dfOrderDetail = dfOrderDetail.rename(columns= {"code" : 'code_order' , 'name' : 'name_order'})[['code_order' , 'name_order' , 'order_id']]
dfManufactureDeptDetail = pd.read_csv('../data_csv/ManufactureDeptDetail.csv')
dfManufactureDeptDetail = dfManufactureDeptDetail.rename(columns= {'code' : 'code_dept' , 'name' : 'name_dept'})
dfManufactureTaskDetail = pd.read_csv('../data_csv/ManufactureTaskDetail.csv')
```

```python
dfManufactureProd = pd.read_csv('../data_csv/ManufactureProd.csv')
dfManufactureProd = dfManufactureProd.rename(columns= {'id' : 'manufacture_product_id' , 'updated_at' : 'create_manu_prod' , 'quantity' : 'quantity_manu_prod' , 'name' : 'name_manu_prod'})
dfManufactureProd = dfManufactureProd.drop(columns= {'created_at','deleted_at' , 'qrcode'})
dfManufactureProd[dfManufactureProd['task_id'] == 1663]
```

```python
# dfManufactureProdImport = pd.read_csv('../data_csv/ManufactureProdImport.csv')
# dfManufactureProdImportDetail = pd.read_csv('../data_csv/ManufactureProdImportDetail.csv')
# dfManufactureProdImportDetail
```

```python
dfManufactureProdImport = pd.read_csv('../data_csv/ManufactureProdImport.csv')
dfManufactureProdImport = dfManufactureProdImport.drop(columns= {'code' , 'deleted_at'})
dfManufactureProdImport[dfManufactureProdImport['task_id'] == 1663]
```

```python
dfManufactureProdImportMerge = dfManufactureProdImport.merge(dfManufactureProd , on = ['manufacture_product_id' , 'task_id'])
dfManufactureProdImportMerge = dfManufactureProdImportMerge.merge(dfOrderDetail , on = 'order_id' , how = 'left')
dfManufactureProdImportMerge = dfManufactureProdImportMerge.merge(dfUser , on = 'creator_id' , how = 'left')
dfManufactureProdImportMerge = dfManufactureProdImportMerge.merge(dfManufactureDeptDetail[['manufacture_dept_id' , 'name_dept']] , on = 'manufacture_dept_id' , how = 'left')
dfManufactureProdImportMerge = dfManufactureProdImportMerge.merge(dfManufactureTaskDetail[['task_id' , 'order_id' , 'start_time' , 'end_time' , 'name_status']])

dfManufactureProdImportMerge[dfManufactureProdImportMerge['task_id'] == 1663].head(2)
```

```python
dfManufactureProdImportFix = dfManufactureProdImportMerge.drop(columns= {'creator_id' , 'created_at' , 'manufacture_dept_id', 'manufacture_product_id' , 'order_id' })

new_col1 = ['task_id', 'code_order' ,'name_order', 'name_dept','fullname', 'updated_at' , 'create_manu_prod' , 'start_time' , 'end_time' ,'quantity_manu_prod' , 
      'quantity', 'position', 'name_status']
   
dfManufactureProdImportFix= dfManufactureProdImportFix.reindex(columns=new_col1)
   
dfManufactureProdImportFix.rename(columns = {'code_order': 'Mã đơn hàng' , 'name_order': 'Tên đơn hàng', 'name_dept':'Phòng ban nhập',  'fullname' : 'Người nhập' , 'updated_at': 'Ngày nhập', 'create_manu_prod' : 'Ngày tạo chi tiết công việc','start_time' : 'Ngày BĐ dự kiến' , 'end_time' : 'Ngày kết thúc dự kiến',
                                             'quantity_manu_prod' : 'Số lượng chi tiết CV' , 'quantity' : 'Số lượng nhập' , 'position' : 'Vị trí' , 
                                             'name_status': 'Trạng thái nhập', 'total_prod_quan': 'Tổng số BTP', 'code_status': 'Mã trạng thái', 'name_status' : 'Trạng thái CV' ,}, inplace = True)
dfManufactureProdImportFix.tail(10)
```

```python
dfManufactureProdImportDetail = pd.read_csv('../data_csv/ManufactureProdImportDetail.csv')
dfManufactureProdImportDetail = dfManufactureProdImportDetail.drop(columns= {'created_at' , 'deleted_at'})
dfManufactureProdImportDetail[dfManufactureProdImportDetail['task_id'] == 3]
```

```python
dfManuProdImportDetailMerge = dfManufactureProdImportDetail.merge(dfManufactureProd , on = ['manufacture_product_id' , 'task_id'])
dfManuProdImportDetailMerge = dfManuProdImportDetailMerge.merge(dfOrderDetail , on = 'order_id' , how = 'left')
dfManuProdImportDetailMerge = dfManuProdImportDetailMerge.merge(dfUser , on = 'creator_id' , how = 'left')
dfManuProdImportDetailMerge = dfManuProdImportDetailMerge.merge(dfManufactureDeptDetail[['manufacture_dept_id' , 'name_dept']] , on = 'manufacture_dept_id' , how = 'left')
dfManuProdImportDetailMerge = dfManuProdImportDetailMerge.merge(dfManufactureTaskDetail[['task_id' , 'order_id' , 'start_time' , 'end_time' , 'name_status']])

dfManuProdImportDetailMerge[dfManuProdImportDetailMerge['task_id'] == 3].head(3)
```

```python
dfManuProdImportDetailMergeFix = dfManuProdImportDetailMerge.drop(columns= {'creator_id' , 'manufacture_dept_id', 'manufacture_product_id' , 'order_id' })

new_col1 = ['task_id', 'code_order' ,'name_order', 'name_dept','fullname', 'updated_at' , 'create_manu_prod' , 'start_time' , 'end_time' ,'quantity_manu_prod' , 
      'quantity', 'position', 'name_status']
   
dfManuProdImportDetailMergeFix= dfManuProdImportDetailMergeFix.reindex(columns=new_col1)
   
dfManuProdImportDetailMergeFix.rename(columns = {'code_order': 'Mã đơn hàng' , 'name_order': 'Tên đơn hàng', 'name_dept':'Phòng ban nhập',  'fullname' : 'Người nhập' , 'updated_at': 'Ngày nhập', 'create_manu_prod' : 'Ngày tạo chi tiết công việc','start_time' : 'Ngày BĐ dự kiến' , 'end_time' : 'Ngày kết thúc dự kiến',
                                             'quantity_manu_prod' : 'Số lượng chi tiết CV' , 'quantity' : 'Số lượng nhập chi tiết' , 'position' : 'Vị trí' , 
                                             'name_status': 'Trạng thái nhập', 'total_prod_quan': 'Tổng số BTP', 'code_status': 'Mã trạng thái', 'name_status' : 'Trạng thái CV' ,}, inplace = True)
dfManuProdImportDetailMergeFix.tail(10)
```

```python
dfManufactureProdImportFix = dfManufactureProdImportFix.sort_values("Ngày kết thúc dự kiến" , ascending=False) 
dfManuProdImportDetailMergeFix = dfManuProdImportDetailMergeFix.sort_values("Ngày kết thúc dự kiến" , ascending=False) 
```

```python
with pd.ExcelWriter('../dataDA/get_manufacture_product_import.xlsx') as writer:  
    dfManufactureProdImportFix.to_excel(writer, sheet_name='import')
    dfManuProdImportDetailMergeFix.to_excel(writer, sheet_name='import_detail')
```
