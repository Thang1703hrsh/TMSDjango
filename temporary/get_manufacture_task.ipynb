```python
import numpy as np
import pandas as pd  
import time
```

```python
dfUser = pd.read_csv('../data_csv/User.csv')
dfOrderDetail = pd.read_csv('../data_csv/OrderDetail.csv')
dfOrderDetail = dfOrderDetail.rename(columns= {"code" : 'code_order' , 'name' : 'name_order'})
```

```python
dfManufactureTask = pd.read_csv('../data_csv/ManufactureTask.csv')
dfManufactureTaskStatus = pd.read_csv('../data_csv/ManufactureTaskStatus.csv')
dfManufactureTaskDetail = pd.read_csv('../data_csv/ManufactureTaskDetail.csv')
dfManufactureTaskLink = pd.read_csv('../data_csv/ManufactureTaskLink.csv')
dfManufactureDeptDetail = pd.read_csv('../data_csv/ManufactureDeptDetail.csv')
dfManufactureDeptDetail = dfManufactureDeptDetail.rename(columns= {'code' : 'code_dept' , 'name' : 'name_dept'})
```

```python
dfUserDel =  dfUser.rename(columns= {"creator_id" : 'user_del_id' , 'fullname' : 'fullname_del'})
dfUserComp =  dfUser.rename(columns= {"creator_id" : 'user_comp_id' , 'fullname' : 'fullname_comp'})

dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfUser, on = 'creator_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfOrderDetail[['code_order' , 'name_order' , 'order_id']], on = 'order_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfManufactureDeptDetail[['code_dept' , 'name_dept' , 'manufacture_dept_id']], on = 'manufacture_dept_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfUserDel, on = 'user_del_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfUserComp, on = 'user_comp_id' , how = 'left')
```
