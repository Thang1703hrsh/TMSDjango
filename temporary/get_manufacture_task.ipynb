```python
import numpy as np
import pandas as pd  
import time
```

```python
dfUser = pd.read_csv('../data_csv/User.csv')
dfOrderDetail = pd.read_csv('../data_csv/OrderDetail.csv')
dfOrderDetail = dfOrderDetail.rename(columns= {"code" : 'code_order' , 'name' : 'name_order'})
dfManufactureTaskStatus = pd.read_csv('../data_csv/ManufactureTaskStatus.csv')
dfManufactureTaskStatus
```

```python
dfManufactureTask = pd.read_csv('../data_csv/ManufactureTask.csv')
dfManufactureTaskStatus = pd.read_csv('../data_csv/ManufactureTaskStatus.csv')
dfManufactureTaskDetail = pd.read_csv('../data_csv/ManufactureTaskDetail.csv')
dfManufactureTaskLink = pd.read_csv('../data_csv/ManufactureTaskLink.csv')
dfManufactureDeptDetail = pd.read_csv('../data_csv/ManufactureDeptDetail.csv')
dfManufactureDeptDetail = dfManufactureDeptDetail.rename(columns= {'code' : 'code_dept' , 'name' : 'name_dept'})
dfManufactureTaskDetail[dfManufactureTaskDetail['task_id'] == 1354]
dfManufactureTaskDetail
# dfManufactureTaskDetail.loc[1794]
```

```python
dfUserDel =  dfUser.rename(columns= {"creator_id" : 'user_del_id' , 'fullname' : 'fullname_del'})
dfUserComp =  dfUser.rename(columns= {"creator_id" : 'user_comp_id' , 'fullname' : 'fullname_comp'})

dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfUser, on = 'creator_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfOrderDetail[['code_order' , 'name_order' , 'order_id']], on = 'order_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfManufactureDeptDetail[['code_dept' , 'name_dept' , 'manufacture_dept_id']], on = 'manufacture_dept_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfUserDel, on = 'user_del_id' , how = 'left')
dfManufactureTaskDetail = dfManufactureTaskDetail.merge(dfUserComp, on = 'user_comp_id' , how = 'left')
```

```python
from bs4 import BeautifulSoup
a = 0

for i in range(0 , dfManufactureTaskDetail.shape[0]):
    aReq = dfManufactureTaskDetail.loc[i]
    soup = BeautifulSoup(aReq['description']).get_text()
    dfManufactureTaskDetail.at[aReq.name , 'description'] = soup
```

```python
dfManufactureTaskDetailFix = dfManufactureTaskDetail.drop(columns= {'code', 'qrcode' , 'creator_id' , 'deleted_at' , 'manufacture_dept_id', 'status_id' , 'order_id' , 'user_del_id' ,
                                                                    'user_comp_id' , 'description_del' })

new_col1 = ['task_id','name' , 'description' , 'code_order' ,'name_order', 'code_dept' ,  'name_dept','fullname', 'created_at', 'updated_at' , 'start_time' , 
      'end_time', 'min_prod_quan', 'total_prod_quan', 'code_status' , 'name_status' , 'fullname_del' , 'fullname_comp']
   
dfManufactureTaskDetailFix= dfManufactureTaskDetailFix.reindex(columns=new_col1)
   
dfManufactureTaskDetailFix.rename(columns = { 'name': 'Tên công việc', 'description': 'Mô tả CV', 'created_at' : 'Ngày tạo' , 'updated_at': 'Ngày Hoàn thành' ,'start_time' : 'Ngày BĐ dự kiến' , 'end_time' : 'Ngày kết thúc dự kiến',
                                             'fullname_del' : 'Nguời xóa CV' , 'fullname_comp' : 'Nguời hoàn tất CV' , 'description_del' : 'Lý do xóa CV' , 
                                             'min_prod_quan': 'Đồng bộ CV', 'total_prod_quan': 'Tổng số BTP', 'code_status': 'Mã trạng thái', 'name_status' : 'Trạng thái CV' ,
                                        'fullname': 'Người tạo CV' , 'code_order': 'Mã đơn hàng' , 'name_order': 'Tên ĐH' , 'code_dept': 'Mã bộ phận',
                                        'name_dept':'Tên bộ phận'}, inplace = True)
dfManufactureTaskDetailFix.tail(10)
```

```python
dfManufactureTaskDetailFix[dfManufactureTaskDetailFix['task_id'] == 1651]
```

```python
dfManufactureTaskDetailComp = dfManufactureTaskDetailFix[dfManufactureTaskDetailFix['Mã trạng thái'] == 'completed']
dfManufactureTaskDetailApproval = dfManufactureTaskDetailFix[(dfManufactureTaskDetailFix['Mã trạng thái'] == 'approval') | (dfManufactureTaskDetailFix['Mã trạng thái'] == 'review')]
dfManufactureTaskDetailImp = dfManufactureTaskDetailFix[dfManufactureTaskDetailFix['Mã trạng thái'] == 'imported']
```

```python
dfManufactureTaskDetailApproval = dfManufactureTaskDetailApproval.sort_values("Ngày kết thúc dự kiến" , ascending=False) 
dfManufactureTaskDetailComp = dfManufactureTaskDetailComp.sort_values("Ngày kết thúc dự kiến" , ascending=False) 
dfManufactureTaskDetailImp = dfManufactureTaskDetailImp.sort_values("Ngày kết thúc dự kiến" , ascending=False) 
```

```python
with pd.ExcelWriter('../dataDA/get_manufacture_task_detail.xlsx') as writer:  
    dfManufactureTaskDetailComp.to_excel(writer, sheet_name='completed')
    dfManufactureTaskDetailApproval.to_excel(writer, sheet_name='approval')
    dfManufactureTaskDetailImp.to_excel(writer, sheet_name='imported')
```
