```python
import numpy as np
import pandas as pd  
import time
```

```python
dfUser = pd.read_csv('../data_csv/User.csv')
dfOrderDetail = pd.read_csv('../data_csv/OrderDetail.csv')
dfOrderDetail = dfOrderDetail.rename(columns= {"code" : 'code_order' , 'name' : 'name_order'})[['code_order' , 'name_order' , 'order_id']]
dfManufactureDeptDetail = pd.read_csv('../data_csv/ManufactureDeptDetail.csv')
dfManufactureDeptDetail = dfManufactureDeptDetail.rename(columns= {'code' : 'code_dept' , 'name' : 'name_dept'})
dfManufactureTaskDetail = pd.read_csv('../data_csv/ManufactureTaskDetail.csv')
```

```python
dfManufactureProd = pd.read_csv('../data_csv/ManufactureProd.csv')
dfManufactureProd = dfManufactureProd.rename(columns= {'id' : 'manufacture_product_id' , 'updated_at' : 'create_manu_prod' , 'quantity' : 'quantity_manu_prod' , 'name' : 'name_manu_prod'})
dfManufactureProd = dfManufactureProd.drop(columns= {'created_at','deleted_at' , 'qrcode'})
dfManufactureProd[dfManufactureProd['task_id'] == 1663]
```

```python
dfManuProdWarehouse = pd.read_csv('../data_csv/ManuProdWarehouse.csv')
dfManuProdWarehouse
```

```python
dfManuProdWarehouseMerge = dfManuProdWarehouse.merge(dfManufactureProd , on = ['manufacture_product_id' , 'task_id'])
dfManuProdWarehouseMerge = dfManuProdWarehouseMerge.merge(dfOrderDetail , on = 'order_id' , how = 'left')
dfManuProdWarehouseMerge = dfManuProdWarehouseMerge.merge(dfUser , on = 'creator_id' , how = 'left')
dfManuProdWarehouseMerge = dfManuProdWarehouseMerge.merge(dfManufactureDeptDetail[['manufacture_dept_id' , 'name_dept']] , on = 'manufacture_dept_id' , how = 'left')
dfManuProdWarehouseMerge = dfManuProdWarehouseMerge.merge(dfManufactureTaskDetail[['task_id' , 'order_id' , 'start_time' , 'end_time' , 'name_status']])

dfManuProdWarehouseMerge[dfManuProdWarehouseMerge['task_id'] == 1663].head(2)
```

```python
dfManuProdWarehouseFix = dfManuProdWarehouseMerge.drop(columns= {'creator_id' , 'id' , 'created_at', 'deleted_at' , 'manufacture_product_id' , 'manufacture_dept_id' , 'order_id', 'create_manu_prod' })
dfManuProdWarehouseFix
```

```python
dfManuProdWarehouseFix = dfManuProdWarehouseMerge.drop(columns= {'creator_id' , 'id' , 'created_at', 'deleted_at' , 'manufacture_product_id' , 'manufacture_dept_id' , 'order_id', 'create_manu_prod' })

new_col1 = ['task_id', 'code_order' ,'name_order', 'position', 'name_manu_prod' , 'name_dept' , 'fullname', 'updated_at' , 'quantity' , 'quantity_manu_prod' , 'start_time' , 'end_time' ,'name_status']
   
dfManuProdWarehouseFix= dfManuProdWarehouseFix.reindex(columns=new_col1)
   
dfManuProdWarehouseFix.rename(columns = {'code_order': 'Mã đơn hàng' , 'name_order': 'Tên đơn hàng', 'name_dept' : 'Phòng ban',  'fullname' : 'Người nhập' , 'updated_at': 'Ngày nhập', 'create_manu_prod' : 'Ngày tạo chi tiết công việc','start_time' : 'Ngày BĐ dự kiến' , 'end_time' : 'Ngày kết thúc dự kiến',
                                             'quantity_manu_prod' : 'Số lượng chi tiết CV' , 'quantity' : 'Số lượng trong kho BTP' , 'position' : 'Vị trí' , 
                                             'name_status': 'Trạng thái'}, inplace = True)
dfManuProdWarehouseFix.tail(10)
```

```python
dfManuProdWarehouseFix.to_excel('../dataDA/get_manufacture_product_warehouse.xlsx')
```
